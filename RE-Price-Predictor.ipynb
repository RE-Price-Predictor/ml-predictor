```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import warnings
```

```python
# Read Cali Income by Zip 
file_path = Path("Resources/NY_ZIP_Income.csv")
df_ny_incomes = pd.read_csv(file_path, header=0)
# Display sample data
df_ny_incomes.head(10)
```

```python
# Read realtor data
file_path = Path("Resources/realtor-data.csv")
df_realtor = pd.read_csv(file_path)
# Display sample data
df_realtor.head(10)
```

```python
joined_df = pd.merge(df_ny_incomes, df_realtor, on='zip_code')
joined_df.head(10)
```

```python
joined_df3 = joined_df.loc[:,["zip_code","income", "status", "bed", "bath", "city", "state", "house_size", "prev_sold_date", "price" ]]
```

```python
joined_df3.head()
```

```python
# Clean the Realtor data
clean_df =  joined_df3.dropna()
clean_df['prev_sold_date'] = pd.to_datetime(clean_df['prev_sold_date'], format="%m/%d/%y") 
clean_df.head()
```

```python
clean_df.dtypes
```

```python
# Creating linear regression model based on the original data that excludes Median Household income by Zip Coded
# Separate the data into labels and features

# Separate the y variable, the labels
y = clean_df["price"]

# Separate the X variable, the features

X = clean_df.drop(columns="price")
#X = X.drop(columns="income")
X = X.drop(columns="status")
X = X.drop(columns="city")
X = X.drop(columns="state")
X = X.drop(columns="prev_sold_date")
```

```python
# Check X table
X.head()
```

```python
clean_df.dtypes
```

```python
# Review the y variable Series
y.head()
```

```python
# Check the balance of our target values
y.value_counts()
```

```python
# Import the train_test_learn module
from sklearn.model_selection import train_test_split

# Split the data using train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

```python
# Import the LinearRegression module from SKLearn
from sklearn.linear_model import LinearRegression

# Instantiate the Linear Regression model
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Predictions

y_pred = regressor.predict(X_test)
```

```python
display(y_pred)
```

```python
pd.DataFrame({"Prediction": y_pred, "Actual": y_test}).tail(10)
```

```python
# Print the balanced_accuracy score of the model
from sklearn.metrics import accuracy_score, classification_report
accuracy_score(y_test, y_pred)
```
