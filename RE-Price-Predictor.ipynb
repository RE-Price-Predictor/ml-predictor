```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import warnings
```

```python
# Read Cali Income by Zip 
file_path = Path("Resources/Cali_Incomes_by_Zip.csv")
df_cali_incomes = pd.read_csv(file_path, header=0)
# Display sample data
df_cali_incomes.head(10)
```

```python
# Clean the data

df_cali_incomes = df_cali_incomes.loc[:, ["Median Household Income", "Zip", "Population"]]
df_cali_incomes.head()
```

```python
# Read realtor data
file_path = Path("Resources/realtor-data.csv")
df_realtor = pd.read_csv(file_path)
# Display sample data
df_realtor.head(10)
```

```python
# Clean the Realtor data

df_realtor = df_realtor.dropna()
display(df_realtor.head())
```

```python
# join the realtor data with Cali incomes by zip based on zip code

joined_data_rows = pd.concat([df_realtor,df_cali_incomes], axis="rows", join="inner")
joined_data_rows
```
